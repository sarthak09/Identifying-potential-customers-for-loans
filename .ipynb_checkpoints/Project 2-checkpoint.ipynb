```python
#Getting X and y
X,y=df.drop(labels='Personal Loan',axis=1).values,df[['Personal Loan']].values
```

```python
df[df.isnull().any(axis=1)]
```

```python
#splitting X and y in tarin and test
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
scale=StandardScaler()
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test) 
```

```python
#Logistic Reg
log_reg=LogisticRegression(random_state=0)
log_reg.fit(X_train,y_train)

y_pred=log_reg.predict(X_test)

confusion_matrix(y_test,y_pred)
```

```python
print("score for train : {0:.4f}".format(log_reg.score(X_train,y_train)))
print("score for test : {0:.4f}".format(log_reg.score(X_test,y_test)))
print("accuracy : {0:.4f}".format(accuracy_score(y_test,y_pred)))
```

```python
print("F1 score : {0:.4f}".format(f1_score(y_test,y_pred)))
print("precision score : {0:.4f}".format(precision_score(y_test,y_pred)))
print("recall score : {0:.4f}".format(recall_score(y_test,y_pred)))
```

```python
print(metrics.classification_report(y_test,y_pred))
```

```python
#Naive bayes
nv_bayes=GaussianNB()
nv_bayes.fit(X_train,y_train)

y_pred=nv_bayes.predict(X_test)

confusion_matrix(y_test,y_pred)
```

```python
print("score for train : {0:.4f}".format(nv_bayes.score(X_train,y_train)))
print("score for test : {0:.4f}".format(nv_bayes.score(X_test,y_test)))
print("accuracy : {0:.4f}".format(accuracy_score(y_test,y_pred)))
```

```python
print("F1 score : {0:.4f}".format(f1_score(y_test,y_pred)))
print("precision score : {0:.4f}".format(precision_score(y_test,y_pred)))
print("recall score : {0:.4f}".format(recall_score(y_test,y_pred)))
```

```python
print(metrics.classification_report(y_test,y_pred))
```

```python
#KNN
knn=KNeighborsClassifier(n_neighbors = 9 , metric='minkowski', p=2) #this is for eucledian only not for minkowski
knn.fit(X_train,y_train)

y_pred=knn.predict(X_test)


confusion_matrix(y_test,y_pred)

```

```python
print("score for train : {0:.4f}".format(knn.score(X_train,y_train)))
print("score for test : {0:.4f}".format(knn.score(X_test,y_test)))
print("accuracy : {0:.4f}".format(accuracy_score(y_test,y_pred)))
```

```python
print("F1 score : {0:.4f}".format(f1_score(y_test,y_pred)))
print("precision score : {0:.4f}".format(precision_score(y_test,y_pred)))
print("recall score : {0:.4f}".format(recall_score(y_test,y_pred)))
```

```python
print(metrics.classification_report(y_test,y_pred))
```

```python
# creating odd list of K for KNN
myList = list(range(1,50))

# subsetting just the odd ones

neighbors = list(filter(lambda x: x % 2 != 0, myList))
```

```python
# empty list that will hold accuracy scores
ac_scores = []

# perform accuracy metrics for values from 1,3,5....19
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # predict the response
    y_pred = knn.predict(X_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
```

```python
#KNN
knn=KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
           metric_params=None, n_jobs=1, n_neighbors=5, p=2,
           weights='uniform')
#knn=KNeighborsClassifier(n_neighbors = 5, metric='minkowski', p=2) #this is for eucledian only not for minkowski
knn.fit(X_train3,y_train3)

y_pred3=knn.predict(X_test3)


confusion_matrix(y_test3,y_pred3)
```

```python
print(knn.score(X_train3,y_train3))
print(knn.score(X_test3,y_test3))
print("accuracy : {0:.4f}".format(accuracy_score(y_test3,y_pred3)))
print("F1 score : {0:.4f}".format(f1_score(y_test3,y_pred3)))
print("precision score : {0:.4f}".format(precision_score(y_test3,y_pred3)))
print("recall score : {0:.4f}".format(recall_score(y_test3,y_pred3)))
print(metrics.classification_report(y_test3,y_pred3))
```
